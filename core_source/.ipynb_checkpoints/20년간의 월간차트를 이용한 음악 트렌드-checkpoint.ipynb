## 20년간 월간차트의 추이변화

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## data항목

```python
names = ['Date', 'Acid/Fusion/POP', 'Drama_OST', 'Classic', 'Indie_Music', 'Movie_OST', 'Ballad', 'Dance', 'Rock/Metal',
         'R&B/Soul', 'Electronica', 'Rap/HipHop', 'EDM', 'POP', 'J-POP', 'Adults_Music', 'Jazz', 'NewAge',
         'Fork/Bules', 'Ani/GamesOST', '-']
```

## final.csv에서 데이터를 가져온다.

```python
def get_data_from_csv():
    # print("current : ", year, month)
    load_file = "C:/Users/jay/Jay/기말플젝/final_data/final.csv"

    try:
        data = pd.read_csv(load_file, encoding='cp949',
                           names=names, header=None)
    except UnicodeError:
        try:
            data = pd.read_csv(load_file, encoding='utf-8',
                               names=names, header=None)
        except Exception as e:
            print(e)
    return data
```

## 데이터(df)를 가지고 그래프를 그리는 함수

```python
def show_graph(df):
    plt.rcParams["figure.figsize"] = (30, 20)
    f = plt.figure()
    df.plot(ax=f.gca())
    plt.xlabel('Year-Month')
    plt.ylabel('Ratio')
    plt.title("Music Trend in 20 years")
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.show()
```

## 사용할 데이터 프레임

```python
mellon_df = get_data_from_csv()
mellon_df
```

## DataFrame 가공

```python
mellon_df['Date'] = pd.to_datetime(mellon_df['Date'], dayfirst=True, errors='coerce')
mellon_df['Date'] = mellon_df['Date']
mellon_df = mellon_df.set_index('Date')
mellon_df = mellon_df.drop(['-'], axis=1, inplace=False)
```

```python
mellon_df.dtypes
```

```python
show_graph(mellon_df)
```
